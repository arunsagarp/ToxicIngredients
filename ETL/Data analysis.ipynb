```python
import pandas as pd
```

```python
df1 = pd.read_csv("../../PubChemLite_31Oct2020_exposomics.csv")
```

```python
df2 = pd.read_csv("../../all_toxin_data.csv")
```

```python
df1.columns
```

```python
df2.columns
```

```python
new_df2 = df2[['common_name','chemical_formula','health_effects']]
```

```python
new_df1 = df1[['MolecularFormula','ToxicityInfo' ]]
```

```python
merged_df = pd.merge(new_df2, new_df1, left_on='chemical_formula', right_on='MolecularFormula', how='outer')
```

```python
total = merged_df.dropna(subset=['common_name', 'ToxicityInfo','health_effects'])
```

```python
total.head(10)
```

```python
sorted_df = total.sort_values(by='ToxicityInfo', ascending=False).reset_index()
```

```python
sorted_df
```

```python
sorted_df = sorted_df[['common_name', 'health_effects']]
```

```python
final_data = sorted_df.drop_duplicates(subset=['common_name'], keep='first')
```

```python
final_data
```

```python
from sqlalchemy import create_engine

# Connect to MySQL database
engine = create_engine('mysql+pymysql://root:root@localhost/ToxicIngredients')


table_name = 'Ingredients'

# Save the dataframe to the MySQL table
final_data.to_sql(table_name, con=engine, if_exists='replace')
```
